# Dot Product of Matrices

Matrix multiplication combines rows of the first matrix with columns
of the second matrix. This operation is fundamental to machine learning.

```python
import numpy as np
```

## Two Matrices

```python
matrix_A = np.array([
    [1, 2],
    [3, 4]
])

matrix_B = np.array([
    [2, 0],
    [1, 2]
])

matrix_A, matrix_B
```

## Matrix Multiplication

```python
result = matrix_A @ matrix_B
result
```

## Why This Matters

Matrix multiplication is used in neural networks,
linear regression, and most machine learning algorithms.
