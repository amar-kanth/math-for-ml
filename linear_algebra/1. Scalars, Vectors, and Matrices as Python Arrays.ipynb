# Import the relevant libraries

```python
import numpy as np # Import the NumPy library, essential for numerical operations in Python, especially with arrays.
```

# Declaring scalars, vectors, and matrices

## Scalars

```python
s = 5 # A scalar is a single numerical value. Here, 's' is assigned the scalar value 5.
```

```python
s # Display the value of the scalar 's'.
```

## Vectors

```python
# A vector is a 1D array of numbers.
v = np.array([5,-2,4]) # By default, a 1D array created with NumPy will behave as a row vector.
```

```python
v # Display the vector 'v'.
```

## Matrices

```python
# A matrix is a 2D array of numbers.
m = np.array([[5,12,6],[-3,0,14]]) # 'm' is a 2x3 matrix with two rows and three columns.
```

```python
m # Display the matrix 'm'.
```

## Data types

```python
type(s) # Check the data type of the scalar 's'. It's a standard Python integer.
```

```python
type(v) # Check the data type of the vector 'v'. It's a NumPy ndarray (n-dimensional array).
```

```python
type(m) # Check the data type of the matrix 'm'. It's also a NumPy ndarray.
```

```python
s_array = np.array(5) # Create a NumPy array from a scalar. This makes it a 0-dimensional NumPy array.
```

```python
type(s_array) # Verify the type of 's_array' is a NumPy ndarray.
```

## Data shapes

```python
m.shape # Get the shape of the matrix 'm'. It returns a tuple (rows, columns).
```

```python
v.shape # Get the shape of the vector 'v'. For a 1D array, it returns a tuple with the number of elements.
```

```python
#s.shape #uncomment to see the error
```

```python
s_array.shape  # it gives empty paranthesis bcz a scalar is a 0-dimentional objects don't have shape.
```

## Creating a column vector

```python
v.reshape(1,3) # Reshape the 1D vector 'v' into a 1x3 row vector (2D array).
```

```python
v.reshape(3,1) # Reshape the 1D vector 'v' into a 3x1 column vector (2D array).
```

```python
m+s # Demonstrates broadcasting: the scalar 's' is added to every element of the matrix 'm'.
```

```python
m+s_array # Similar to adding a scalar, the 0-dimensional NumPy array 's_array' is broadcasted to 'm'.
```

```python
v+s # The scalar 's' is broadcasted and added to each element of the vector 'v'.
```

```python
v # Display the original vector 'v' to show it remains unchanged.
```

```python
v+s_array # The 0-dimensional NumPy array 's_array' is broadcasted and added to each element of the vector 'v'.
```

```python
m+v # Demonstrates broadcasting: the vector 'v' is added row-wise to the matrix 'm'.
```
