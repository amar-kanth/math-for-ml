# Import the relevant libraries

```python
import numpy as np # Import the numpy library for numerical operations, especially with arrays and matrices.
```

## Addition

```python
m1 = np.array([[5,12,6],[-3,0,14]]) # Define the first 2x3 matrix m1
m1
```

```python
m2 = np.array([[9,8,7],[1,3,-5]]) # Define the second 2x3 matrix m2
m2
```

```python
m1 + m2 # Perform element-wise addition of matrix m1 and m2
```

## Difference

### Example 1

```python
m3 = np.array([[5,3],[-2,4]]) # Define the first 2x2 matrix m3
m4 = np.array([[7,-5],[3,8]]) # Define the second 2x2 matrix m4
```

```python
m3 # Display matrix m3
```

```python
m4 # Display matrix m4
```

```python
m3 - m4 # Perform element-wise subtraction of matrix m4 from m3
```

## Adding vectors together

```python
v1 = np.array([1,2,3,4,5]) # Define the first 1D array (vector) v1
v2 = np.array([5,4,3,2,1]) # Define the second 1D array (vector) v2
```

```python
v1 + v2 # Perform element-wise addition of vector v1 and v2
```

```python
v1 - v2 # Perform element-wise subtraction of vector v2 from v1
```
