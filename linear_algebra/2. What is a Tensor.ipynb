# Import the relevant libraries

```python
import numpy as np # Import the numpy library, commonly aliased as 'np', for numerical operations.
```

## Creating a tensor

```python
m1 = np.array([[5,12,6],[-3,0,14]]) # Create a 2x3 NumPy array (matrix) named m1.
m1 # Display the matrix m1.
```

```python
m2 = np.array([[9,8,7],[1,3,-5]]) # Create another 2x3 NumPy array (matrix) named m2.
m2 # Display the matrix m2.
```

```python
t = np.array([m1,m2]) # Create a 3D NumPy array (tensor) by stacking m1 and m2 along a new axis.
```

```python
t # Display the tensor t.
```

## Checking its shape

```python
t.shape # Get and display the shape of the tensor t. It is a 2x2x3 object, meaning it contains 2 matrices of 2x3 each.
```

## Manually creating a tensor

```python
t_manual = np.array([[[ 5, 12,  6], [-3,  0, 14]], [[ 9,  8,  7], [ 1,  3, -5]]]) # Manually create a 3D NumPy array (tensor) with specified values.
```

```python
t_manual # Display the manually created tensor t_manual.
```
