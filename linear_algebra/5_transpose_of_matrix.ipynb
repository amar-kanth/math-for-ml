# Transpose of a Matrix

The transpose of a matrix is obtained by swapping its rows and columns.
This operation is widely used in linear algebra and machine learning.

```python
import numpy as np
```

## Original Matrix

```python
data_matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

data_matrix
```

## Transposed Matrix

```python
transposed_matrix = data_matrix.T
transposed_matrix
```

## Why Transpose is Useful

Transpose is often used when switching between row-wise and column-wise representations,
especially in machine learning algorithms.

## Machine Learning Connection

Transpose is used when switching between:
- Features × Samples
- Samples × Features

```python
X = np.array([
    [1, 2],
    [3, 4],
    [5, 6]
])

X, X.T
```
