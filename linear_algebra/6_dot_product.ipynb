# Dot Product of Vectors

The dot product is one of the most important operations in linear algebra.
It measures how similar two vectors are and is used heavily in machine learning.

```python
import numpy as np
```

## Computing the Dot Product

```python
vector_a = np.array([2, 3, 4])
vector_b = np.array([5, 6, 7])

dot_result = np.dot(vector_a, vector_b)
dot_result
```

## Manual Interpretation

The dot product is calculated as:
a₁b₁ + a₂b₂ + a₃b₃

```python
manual_dot = (2 * 5) + (3 * 6) + (4 * 7)
manual_dot
```

Both results should be the same.
