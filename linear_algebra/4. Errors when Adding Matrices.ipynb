### Addition of matrices

```python
# Define a 2x3 matrix m1 using NumPy array
m1 = np.array([[5,12,6],[-3,0,14]])
m1
```

```python
# Define a 2x2 matrix m3 using NumPy array
m3 = np.array([[5,3],[-2,4]])
m3
```

### Addition of vectors

```python
# Define a vector v1 using NumPy array
v1 = np.array([1,2,3,4,5])
v1
```

```python
# Define a vector v3 using NumPy array
v3 = np.array([1,2,3])
v3
```
